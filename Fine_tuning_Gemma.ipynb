## LoRA Fine-tuning

```
# Enable LoRA for the model and set the LoRA rank to 4.
gemma_lm.backbone.enable_lora(rank=lora_rank)
gemma_lm.summary()

# Limit the input sequence length (to control memory usage).
gemma_lm.preprocessor.sequence_length = token_limit
# Use AdamW (a common optimizer for transformer models).
optimizer = keras.optimizers.AdamW(
    learning_rate=lr_value,
    weight_decay=0.01,
)
# Exclude layernorm and bias terms from decay.
optimizer.exclude_from_weight_decay(var_names=["bias", "scale"])

gemma_lm.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=optimizer,
    weighted_metrics=[keras.metrics.SparseCategoricalAccuracy()],
)
```

```
class CustomCallback(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs=None):
    model_name = f"./gemma_workshop/{lora_name}_{lora_rank}_epoch{epoch+1}.lora.h5"
    gemma_lm.backbone.save_lora_weights(model_name)

    # Evaluate
    text_gen("Write a title")
    text_gen("Write a poem")
```

```
import tensorflow as tf

from datasets import load_dataset

dataset = load_dataset("nbertagnolli/counsel-chat", split="train")

def map_to_dict(example):
    """Maps dataset examples to a dictionary with 'prompts' and 'responses' keys, handling None values."""
    # Replace None values with empty strings in both columns
    prompt = example["questionText"] if example["questionText"] is not None else ""
    response = example["answerText"] if example["answerText"] is not None else ""

    return {
        "prompts": tf.constant(prompt, dtype=tf.string),
        "responses": tf.constant(response, dtype=tf.string)
    }

# Apply the mapping function to your dataset and convert to TensorFlow Dataset
train_dataset = dataset.map(map_to_dict)
train_dataset = train_dataset.to_tf_dataset(
    columns=["prompts", "responses"],
    batch_size=2,
    shuffle=True,
)

# Now, use train_dataset for fitting the model
history = gemma_lm.fit(train_dataset, epochs=train_epoch, callbacks=[CustomCallback()])
```
